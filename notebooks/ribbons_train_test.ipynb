# ribbons - train test valid

finalne stahovanie obrazkov do train test valid pre triedy C, M a X (ribbondb_v1.0.csv)

ribbonsb_v1.0.csv 

5 pred + PEAK + 5 po

```python
#final final final final final final
import os
import pandas as pd
import hvpy
from hvpy.datasource import DataSource
from datetime import datetime, timedelta
from sunpy.map import Map
from sunpy.coordinates import frames
from astropy.coordinates import SkyCoord
import astropy.units as u
from sklearn.model_selection import train_test_split

# priecinky 
base_folder = "./ribbons"
train_folder = os.path.join(base_folder, "train")
test_folder = os.path.join(base_folder, "test")
valid_folder = os.path.join(base_folder, "validation")
class_folders = ['C', 'M', 'X']

# priecinky pre cele obr a vyrezy
for folder in [train_folder, test_folder, valid_folder]:
    for cls in class_folders:
        os.makedirs(os.path.join(folder, cls), exist_ok=True)

# nacitanie datasetu
data = pd.read_csv("./csv's/ribbondb_v1.0.csv")
data_sources = hvpy.getDataSources()
source_id = data_sources['SDO']['AIA']['1600']['sourceId']
data['event_id'] = data[' KEY'].str[:-5]  # extrahovanie triedy z atributu key
unique_events = data['event_id'].unique()

# rozdelenie do train, test a valid v pomere train:test - 8:2, test:valid - 9:1
train_events, test_events = train_test_split(unique_events, test_size=0.2, random_state=42)
train_events, valid_events = train_test_split(train_events, test_size=0.1, random_state=42)

# funkcia na stahovanie obrazkov a vytvorenie vyrezov
def download_images(event_list, base_target_folder):
    for index, row in data[data['event_id'].isin(event_list)].iterrows():
        key = row[' KEY']
        event_class = key[-4]  # trieda (C, M, X)
        if event_class not in class_folders:
            continue  # preskoci ak nie je C, M, or X

        # casy udalosti vo formate rok-mesiac-den hodina:minuta
        tpeak = datetime.strptime(row[' TPEAK [UT]'].strip(), '%Y-%m-%d %H:%M')
        tstart = datetime.strptime(row[' TSTART [UT]'].strip(), '%Y-%m-%d %H:%M')
        tfinal = datetime.strptime(row[' TFINAL [UT]'].strip(), '%Y-%m-%d %H:%M')
        duration = (tfinal - tstart).total_seconds() / 60

        # pre udalosti, ktorych trvanie bolo dlhsie ako 11 minut sme vybrali snimky, ktore boli 5 minut pred/po case TPEAK
        if duration >= 11:
            times = [tpeak - timedelta(minutes=5) + timedelta(minutes=i) for i in range(11)]
        else:
            times = pd.date_range(tstart, tfinal, freq='min').to_pydatetime().tolist()

        # suradnice pre vyrezy
        img_lat = row[' LAT [deg]']
        img_lon = row[' LON [deg]']
        top_right = SkyCoord((img_lon - 20) * u.deg, (img_lat + 20) * u.deg, frame=frames.HeliographicStonyhurst)
        bottom_left = SkyCoord((img_lon + 20) * u.deg, (img_lat - 20) * u.deg, frame=frames.HeliographicStonyhurst)

        # stahovanie a vyrezy
        for time in times:
            date_str = time.strftime('%Y-%m-%dT%H:%M')
            filename_suffix = f"{time.strftime('%Y%m%d_%H%M')}_{event_class}"
            jp2_filename = f"{filename_suffix}.jp2"
            cutout_filename = f"{filename_suffix}_cutout.fits"
            jp2_path = os.path.join(base_target_folder, event_class, jp2_filename)
            cutout_path = os.path.join(base_target_folder, event_class, cutout_filename)

            # preskoci ak subor existuje
            if os.path.exists(jp2_path) or os.path.exists(cutout_path):
                print(f"Skipping: {filename_suffix} (already exists)")
                continue

            try:
                # stahovanie pomocou hvpy
                image_content = hvpy.getJP2Image(date=date_str, sourceId=source_id)
                with open(jp2_path, 'wb') as f:
                    f.write(image_content)
                print(f"Downloaded: {jp2_path}")
            except Exception as e:
                print(f"Error downloading {date_str}: {e}")
                continue

            # vytvorenie vyrezu
            try:
                file = Map(jp2_path)
                aia_sub = file.submap(bottom_left, top_right=top_right)
                aia_sub.save(cutout_path, overwrite=True)
                os.remove(jp2_path)  # vymazanie celeho JP2 obrazku
                print(f"Cutout saved to: {cutout_path}")
            except Exception as e:
                print(f"Error creating cutout {cutout_path}: {e}")
                if os.path.exists(jp2_path):
                    os.remove(jp2_path)  # zabezpecenie ze JP2 obrazok sa odstranil
                continue

# stahovanie a vytvorenie vyrezov pre train, test, and valid (debug)
print("Downloading train images...")
download_images(train_events, train_folder)

print("Downloading validation images...")
download_images(valid_events, valid_folder)

print("Downloading test images...")
download_images(test_events, test_folder)

print("All images downloaded and cutouts created.")
```
