# Štatistiky trvania udalostí

```python
# uistenie že všetky hodnoty sú reťazce a odstránenie medzier
data[' TSTART [UT]'] = data[' TSTART [UT]'].astype(str).str.strip()
data[' TFINAL [UT]'] = data[' TFINAL [UT]'].astype(str).str.strip()

# konverzia časových stĺpcov na formát datetime
data[' TSTART [UT]'] = pd.to_datetime(data[' TSTART [UT]'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
data[' TFINAL [UT]'] = pd.to_datetime(data[' TFINAL [UT]'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# výpočet trvania udalosti (v sekundách) pre každý riadok a následne prepočet na minúty
data['Duration (min)'] = (data[' TFINAL [UT]'] - data[' TSTART [UT]']).dt.total_seconds() / 60

# vypísanie základných štatistík pre stĺpec trvania v minútach
statistics = {
    'Min': data['Duration (min)'].min(),
    'Max': data['Duration (min)'].max(),
    'Mean': data['Duration (min)'].mean(),
    'Median': data['Duration (min)'].median(),
}

print("štatistiky trvania udalostí (v minútach):")
for stat, value in statistics.items():
    print(f"{stat}: {value}")
```

```python
data[' TSTART [UT]'] = data[' TSTART [UT]'].astype(str).str.strip()
data[' TFINAL [UT]'] = data[' TFINAL [UT]'].astype(str).str.strip()

# konverzia časových stĺpcov na formát datetime
data[' TSTART [UT]'] = pd.to_datetime(data[' TSTART [UT]'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
data[' TFINAL [UT]'] = pd.to_datetime(data[' TFINAL [UT]'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# výpočet trvania udalosti (v sekundách) pre každý riadok a následne prepočet na minúty
data['Duration (s)'] = (data[' TFINAL [UT]'] - data[' TSTART [UT]']).dt.total_seconds() 

# vypísanie základných štatistík pre stĺpec trvania v minútach
statistics = {
    'Min': data['Duration (s)'].min(),
    'Max': data['Duration (s)'].max(),
    'Mean': data['Duration (s)'].mean(),
    'Median': data['Duration (s)'].median(),
}

print("štatistiky trvania udalostí (v sekundach):")
for stat, value in statistics.items():
    print(f"{stat}: {value}")
```

```python
import os
import random
import pandas as pd
import hvpy
from sunpy.map import Map
from astropy.coordinates import SkyCoord
import astropy.units as u
from sunpy.coordinates import frames
from datetime import datetime
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np

# nacitanie datasetu
data = pd.read_csv("./csv's/ribbondb_v1.0.csv")

# nacitaniee zdroja pre AIA 1600
data_sources = hvpy.getDataSources()
source_id = data_sources['SDO']['AIA']['1600']['sourceId']

# cielovy priecinok kde sa ulozia obr
output_dir = "outputsData"
os.makedirs(output_dir, exist_ok=True)

# vyber nahodnej udalosti z tried (C, M, X)
selected_rows = []
for flare_class in ['C', 'M', 'X']:
    class_rows = data[data[' KEY'].str[-4] == flare_class]
    if not class_rows.empty:
        selected_rows.append(class_rows.sample(1).iloc[0])

# funkcia na zisknie dat a vyrezu
def get_cutout_image(row, time_point):
    img_lat = row[' LAT [deg]']
    img_lon = row[' LON [deg]']
    key = row[' KEY']
    time_str = row[time_point].strip()
    time = datetime.strptime(time_str, "%Y-%m-%d %H:%M")
    date_str = time.strftime('%Y-%m-%dT%H:%M')

    top_right = SkyCoord((img_lon - 20) * u.deg, (img_lat + 20) * u.deg, frame=frames.HeliographicStonyhurst)
    bottom_left = SkyCoord((img_lon + 20) * u.deg, (img_lat - 20) * u.deg, frame=frames.HeliographicStonyhurst)

    try:
        image_content = hvpy.getJP2Image(date=date_str, sourceId=source_id)
        with open("temp.jp2", 'wb') as f:
            f.write(image_content)
        m = Map("temp.jp2")
        cutout = m.submap(bottom_left, top_right=top_right)
        return cutout.data, time.strftime("%Y-%m-%d %H:%M"), key
    except Exception as e:
        print(f"chyba: {e}")
        return None, None, None   

# vykreslenie a ulozenie
for row in selected_rows:
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))
    fig.suptitle(row[' KEY'], fontsize=14, color='blue')

    filenames = []

    for i, col in enumerate([" TSTART [UT]", " TPEAK [UT]", " TFINAL [UT]"]):
        img_data, label, key = get_cutout_image(row, col)
        if img_data is not None:
            axes[i].imshow(img_data, cmap='gray')
            axes[i].set_title(label)
        else:
            axes[i].text(0.5, 0.5, "obr nedostupny", ha='center', va='center')
        axes[i].axis('off')

    plt.tight_layout()
    plt.subplots_adjust(top=0.8)
    plt.show()
    # nazov suboru podla KEY
    safe_key = row[' KEY'].replace(" ", "").replace(":", "").replace(".", "")
    save_path = os.path.join(output_dir, f"{safe_key}.png")
    plt.savefig(save_path, bbox_inches='tight', dpi=150)
    plt.close()
    print(f"obr saved: {save_path}")
```

# Graf pre počet obrázkov v pričinkoch train, test a valid jednotlivých tried z datasetu ribbondb_v1.0.csv (ribbons)

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# priecinky
base_path = "./ribbons"
splits = ["train", "test", "valid"]
classes = ["C", "M", "X"]
colors = {'C': '#2A4494', 'M': '#72A1E5', 'X': '#9BD1E5'}

# ziskavanie poctov obr
counts = []
for split in splits:
    for cls in classes:
        folder = os.path.join(base_path, split, cls)
        count = len([f for f in os.listdir(folder) if f.endswith(".png")])
        counts.append((split, cls, count))

# priprava dat
labels = []
values = []
bar_colors = []
group_positions = {'train': 0, 'test': 4, 'valid': 8}
bar_positions = []

for i, (split, cls, count) in enumerate(counts):
    base_pos = group_positions[split]
    offset = classes.index(cls)
    bar_positions.append(base_pos + offset)
    values.append(count)
    labels.append(split)
    bar_colors.append(colors[cls])

# vykreslenie grafu
plt.figure(figsize=(10, 6))
bars = plt.bar(bar_positions, values, color=bar_colors)

# pocty nad stlpcami
for bar, value in zip(bars, values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height * 1.05, f'{value}', ha='center', va='bottom', fontsize=8)

# popis pre train test a valid
plt.xticks([1, 5, 9], ["Trénovacia množina", "Testovacia množina", "Validačná množina"])

# legenda
legend_labels = [plt.Rectangle((0,0),1,1, color=colors[cls]) for cls in classes]
plt.legend(legend_labels, classes, title='Trieda')

plt.yscale("log")
plt.ylabel("Počet obrázkov")
#plt.title("Počet obrázkov pre jednotlivé triedy (C, M a X) rozdelených na trénovaciu, testovaciu a validačnú množinu")
plt.tight_layout()

# ulozenie grafu
output_path = "outputsData/pocetTrainTestValidRibbons.png"  # cesta a nazov suboru
plt.savefig(output_path, bbox_inches='tight')  # ulozi obrazok

plt.show()
```
